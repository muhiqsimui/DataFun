Numpy Arange ( nilai floating awal, nilai floating max, interval )

```python
t=np.arange(0.,5.,0.2)
t
```

```python
plt.plot(t,t,'r--') 
plt.plot(t,t**2,'bs') # s = kotak
plt.plot(t,t**3,'g^') # ^ = segitiga
plt.show()
```

```python
plt.plot(t,t,'r--',t,t**2,'bs',t,t**3,'g^')
plt.title('multi plot')
plt.show()
```

- **np.arange(jumlah)** panggil 50 bilangan
- **np.random.randint(value awal, value akhir, jumlah)** acak bilangan dalam rentang value awal sampai akhir sebanyak jumlah
- **np.random.randn(jumlah)** acak sejumlah 50 saja

```python
data={
    'a':np.arange(50),
    'b':np.random.randint(0,50,50),
    'c':np.random.randn(50)
}

data
```

```python
data['d'] = data['a']+10*np.random.randn(50)
data['c'] = np.abs(data['d'])*100
data
```

```python
plt.scatter('a','d',data=data)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title("Ok")
plt.show()
```

**c** itu color atau warna, **s** itu sizenya atau ukuran


```python
plt.scatter('a','d',c='b',s='c',data=data)
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.show()
```
