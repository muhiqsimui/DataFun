Ada 2 Cara visualisasi dengan matplotlib
- Object Oriented Style
- pylot Style

**1. Object Oriented Style**

Membuat figure serta axes, dan memanggil method dari keduanya

```python
# Numpy linspace untuk melakukan generate terhadap sejumlah deret bilangan
# np.linspace( Nilai awal , Nilai Akhir , Jumlah data point yang ingin dibuat)
x = np.linspace(0, 2, 100)
# x = np.linspace(1,10,100)
print(x)
```

```python
fig,ax=plt.subplots()

# sebenarnya isi dari method plot tetap x, y hanya saja datanya di generate menggunakan linspace jadi yang dipanggil x ,x sama aja kok
ax.plot(x, x, label='Linear') # ini linearnya
ax.plot(x, x**2, label="Quadratic") # dipangkatkan 2
ax.plot(x, x**3, label="Cubic") # dipangkatkan 3
#label sudut x dan sudut y
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
#ini buat judul diatasnya
ax.set_title('Plotting Sederhana')
#nampilin kotak kecil buat label
ax.legend()
```

**2. Pyplot Style**

Dimana kita akan mengandalkat Pyplot untuk membuat dan mengelola figures dan axes, serta menggunakan fungsi pyplot untuk melakukan plotting


```python
#tidak perlu membuat figure dan axes fig,ax=subplots() karena dibuatkan pyplot secara otomatis

#generate data set
x=np.linspace(0,2,100)

plt.plot(x,x,label='linear')
plt.plot(x,x**2,label='kuadrat')
plt.plot(x,x**3,label='kubik')

#tidak pake set_xlabel langsung xlabel saja kalo pyplot
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')
plt.title('Plotting Pyplot Style')
plt.legend()
```
